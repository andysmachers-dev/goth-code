```python
import numpy as np
import matplotlib.pyplot as plt

# Create a range of input values
x = np.linspace(-10, 10, 100)

# --- 1. Define Activation Functions ---

def sigmoid(x):
    """Sigmoid activation function"""
    return 1 / (1 + np.exp(-x))

def tanh(x):
    """Tanh activation function"""
    return np.tanh(x)

def relu(x):
    """ReLU activation function"""
    return np.maximum(0, x)

# --- 2. Define Their Derivatives (Gradients) ---

def sigmoid_derivative(x):
    """Derivative of the sigmoid function"""
    s = sigmoid(x)
    return s * (1 - s)

def tanh_derivative(x):
    """Derivative of the tanh function"""
    t = tanh(x)
    return 1 - t**2

def relu_derivative(x):
    """Derivative of the ReLU function"""
    # 1 if x > 0, 0 otherwise
    return np.where(x > 0, 1, 0)

# --- 3. Plotting ---

# Set up the plots
plt.figure(figsize=(14, 6))

# Subplot 1: The Activation Functions
plt.subplot(1, 2, 1)
plt.plot(x, sigmoid(x), label='Sigmoid', color='blue')
plt.plot(x, tanh(x), label='Tanh', color='green')
plt.plot(x, relu(x), label='ReLU', color='red')
plt.title('Activation Functions')
plt.xlabel('Input (x)')
plt.ylabel('Output (f(x))')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Subplot 2: The Gradients (Derivatives)
plt.subplot(1, 2, 2)
plt.plot(x, sigmoid_derivative(x), label='Sigmoid Derivative', linestyle='--', color='blue')
plt.plot(x, tanh_derivative(x), label='Tanh Derivative', linestyle='--', color='green')
plt.plot(x, relu_derivative(x), label='ReLU Derivative', linestyle='--', color='red')
plt.title('Derivatives (Gradients)')
plt.xlabel('Input (x)')
plt.ylabel('Gradient (f\'(x))')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Show the plots
plt.tight_layout()
plt.show()
```
